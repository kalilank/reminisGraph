```python
import pandas as pd
from pathlib import Path
```

```python
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

vector_path = PROJECT_ROOT / "data" / "processed" / "vector_retrieval_results.csv"
graph_path = PROJECT_ROOT / "data" / "processed" / "graph_retrieval_results.csv"

vector_results = pd.read_csv(vector_path)
graph_results = pd.read_csv(graph_path)

vector_results.head(), graph_results.head()
```

```python
print("Vector columns:", vector_results.columns.tolist())
print("Graph columns:", graph_results.columns.tolist())
```

```python
def compare_artist(query_artist):
    vector_view = vector_results[
        vector_results["query_artist"] == query_artist
    ].copy()

    vector_view["retrieval_type"] = "vector"
    vector_view["score"] = vector_view["similarity_score"]
    vector_view["evidence_type"] = "-"
    vector_view["evidence"] = "-"

    vector_view = vector_view[
        [
            "query_artist",
            "retrieval_type",
            "rank",
            "retrieved_artist",
            "score",
            "evidence_type",
            "evidence"
        ]
    ]

    graph_view = graph_results[
        graph_results["query_artist"] == query_artist
    ].copy()

    graph_view["retrieval_type"] = "graph"
    graph_view["score"] = graph_view["graph_score"]

    graph_view = graph_view[
        [
            "query_artist",
            "retrieval_type",
            "rank",
            "retrieved_artist",
            "score",
            "evidence_type",
            "evidence"
        ]
    ]

    comparison = pd.concat([vector_view, graph_view], ignore_index=True)

    return comparison
```

#### Phase 9A Conclusion

This notebook compares the vector retrieval baseline and the graph retrieval baseline.

The vector retriever ranks artists based on embedding similarity from album and tag text. Its score is a cosine similarity value, where higher values mean the artist embeddings are more similar.

The graph retriever ranks artists based on explicit MusicBrainz graph evidence, such as shared members and shared tags. Its graph score is not a normalized similarity score. Instead, it represents the number of graph evidence items found for each relationship type.

Because of this, vector scores and graph scores should not be interpreted on the same scale. A vector score close to 1 means high embedding similarity, while a graph score of 1 shared member can still be meaningful because it represents direct historical relationship evidence.

One important example is Pearl Jam. The vector retriever ranks Blur highest for Pearl Jam based on text/tag similarity, while the graph retriever surfaces Soundgarden through Matt Cameron as shared-member evidence. This shows that graph retrieval can reveal historical or relational music connections that may not be ranked highest by vector similarity alone.

Overall, this comparison supports the main project idea: vector retrieval is useful for style/tag similarity, while graph retrieval is useful for relationship-based discovery.

The next step is to design a simple hybrid retrieval approach that combines vector similarity with graph-based evidence.
