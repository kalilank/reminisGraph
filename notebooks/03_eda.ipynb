## Album-level EDA Insight

The dataset currently contains 6 selected studio albums from 6 seed artists, with 70 total recordings/tracks. Each album contains between 10 and 13 tracks, which suggests that the release selection strategy produced clean album-level data rather than noisy deluxe editions or compilations.

Album duration ranges from around 42 to 50 minutes. Pearl Jam's `Ten` has one missing track length, so its total duration is slightly underestimated.

For the future graph schema, this EDA supports a basic hierarchical structure:

Artist -> ReleaseGroup/Album -> Release -> Recording

At this stage, `release_date`, `country`, `medium_format`, `track_position`, and `length_min` are better modeled as properties rather than separate nodes.

## Release Edition Metadata Check

The `country` and `medium_format` fields appear to describe the selected MusicBrainz release edition, not the artist's origin or the album's cultural/geographic origin.

For example, `Bleach` by Nirvana is associated with `CA`, and `Gish` by The Smashing Pumpkins is associated with `NL`. These values should not be interpreted as artist origin countries.

Therefore, in the future graph schema, `country` and `medium_format` should be modeled as properties of the `Release` node, not as separate `Country` nodes connected to `Artist`.

Candidate modeling decision:

Artist -> ReleaseGroup/Album -> Release -> Recording

Release properties:
- release_date
- country
- medium_format

## Relationship Richness Check

The current `album_tracks.csv` mainly represents a hierarchical music structure:

Artist -> ReleaseGroup/Album -> Release -> Recording

From the available columns, the dataset can describe which tracks belong to which release and artist. However, it does not yet contain richer cross-artist relationships such as shared members, shared producers, shared labels, tags/genres, or artist-to-artist relationships.

This means the current CSV is useful for validating the ingestion pipeline and designing the basic graph structure, but it is not yet enough to fully demonstrate why graph retrieval may outperform vector retrieval.

## Phase 3 Interim Conclusion

The current dataset is clean and structurally consistent. It contains 6 seed artists, 6 selected studio albums, and 70 unique recordings/tracks.

Genre/tag distribution is not analyzed in this phase because `album_tracks.csv` does not yet include genre or tag fields. Tags are planned as candidate relationship-rich metadata for the next graph design phase.

The EDA supports a basic graph schema:

Artist → ReleaseGroup/Album → Release → Recording

However, this structure is mostly hierarchical and does not yet capture the relationship-rich data that makes Knowledge Graph retrieval valuable. For Phase 4, the graph design should incorporate richer MusicBrainz relationships such as artist relationships, member-of relationships, label relationships, and possibly tags/genres.

Therefore, the next design question is not only “How do we represent albums and tracks?”, but “What relationships allow graph traversal to discover connections that vector retrieval may miss?”
